```python
import numpy as np
import pandas as pd
from numpy.random import randn

#stats lib from scipy
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#command so that plots appear in iPython Notebook
%matplotlib inline
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)

plt.ylim(0,1) #set y axis scale
```

```python
plt.hist(dataset, alpha=0.3)
sns.rugplot(dataset)
```

```python
sns.kdeplot(dataset)
```

```python
sns.rugplot(dataset, color='black')

for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset, bw=bw, lw=1.8, label=bw)
```

```python
# kernel options 
kernel_options = ['biw', 'cos', 'epa', 'gau', 'tri', 'triw']

for kern in kernel_options:
    sns.kdeplot(dataset, kernel=kern, label=kern, shade=True)
```

```python
#plot on Y axis
sns.kdeplot(dataset,vertical=True)
```

```python
# CDF - Cumulative Distibution Function
sns.kdeplot(dataset, cumulative=True)
```

```python
# Multivariant Density Estimation
mean = [0,0]
cov = [[1,0], [0,100]]

dataset2 = np.random.multivariate_normal(mean, cov, 100)

dframe = pd.DataFrame(dataset2, columns=['X','Y'])

sns.kdeplot(dframe)
```

```python
# KDE joint plot
sns.jointplot('X','Y',dframe, kind='kde')
```
