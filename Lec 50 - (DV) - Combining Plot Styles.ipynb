```python
import numpy as p
import pandas as pd
from numpy.random import randn

#stats lib from numpy
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#plots to appear in iPython notebook
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
sns.distplot(dataset,bins=25) #combines KDE with histogram
```

```python
sns.distplot(dataset,bins=25,rug=True, hist=False) # chaning default histogram with rug plot 
```

```python
#controlling properties of combined plots by passing dict
sns.distplot(dataset, bins=25,
             kde_kws={'color':'indianred','label':'KDE PLOT'},
             hist_kws={'color':'blue', 'label':'HIST'}
            )
```

```python
# use Pandas with Seaborn
from pandas import Series
ser1 = Series(dataset, name="My_Data")
```

```python
sns.distplot(ser1)
```
