```python
#Normal imports
import numpy as np
import pandas as pd
from numpy.random import randn

#stats
from scipy import stats

#plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#to show plots in iPyNotebook
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)

sns.boxplot([data1, data2])
```

```python
#include outliers in whisker i.e. set min max/whisker to infinity
sns.boxplot([data1, data2], whis=np.inf)
```

```python
#vertical boxplot i.e. set orientation
sns.boxplot([data1, data2], orient='v')
```

```python
# Normal Distribution
data1 = stats.norm(0,5).rvs(100) # get 100 points from distribution

#2 gamma distributions & concatenate together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1, 
                      -1*stats.gamma(5).rvs(50)])

#Box plot both data 1 & data 2
sns.boxplot([data1, data2], whis=np.inf, orient='v')
```

```python
# Violin Plot = KDE + Box Plot

sns.violinplot([data1, data2], orient='v')
```

```python
# change bandwidth
sns.violinplot(data2, bw=0.02)
```

```python
#add rug plot idea to violin plot
sns.violinplot(data1, inner='stick')
```
